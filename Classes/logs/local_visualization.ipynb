```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

data_type = 'kaggle'
reward_type = 'probability'
# 
# Load the CSV file
file_path = '2024-10-14-17-17-17/'+data_type+'/_1/balance=0.1/1/k-4/u-6/fraudulent/reward-' + str(reward_type)+'/file.csv'
data = pd.read_csv(file_path)

# Set the moving average window size (adjust as needed)
window_size = 300
moving_average = data.rolling(window=window_size).mean()

# Plot each column's moving average with its name as the label
plt.figure(figsize=(10, 6))

# Loop through each column and plot its moving average
for column in moving_average.columns:
    plt.plot(moving_average[column], label=column)

# Add labels and legend
plt.xlabel('Frauds')
plt.ylabel('Reward')
plt.title(f'Reward over time: ' + str(reward_type) + ' Data type: ' + str(data_type))
plt.legend()
plt.show()

print(f'Moving Average with Window Size {window_size}')
```
