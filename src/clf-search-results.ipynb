```python
from datetime import datetime, timedelta
from banksys import Transaction, Card, Terminal, Banksys
import polars as pl

from parameters import ClassificationParameters

cards = pl.DataFrame([Card(0, 10, 25, 500), Card(1, 20, 30, 1000)])
terminals = pl.DataFrame([Terminal(0, 75, 95), Terminal(1, 17, 56)])
transactions = pl.DataFrame(
    [
        # Warmup
        Transaction(100, datetime(2023, 1, 1), terminal_id=0, card_id=0, is_online=False, is_fraud=False),
        Transaction(200, datetime(2023, 1, 2), terminal_id=1, card_id=1, is_online=True, is_fraud=False),
        Transaction(150, datetime(2023, 1, 2), terminal_id=1, card_id=1, is_online=True, is_fraud=False),
        Transaction(120, datetime(2023, 1, 5), terminal_id=0, card_id=0, is_online=False, is_fraud=False),
        Transaction(180, datetime(2023, 1, 10), terminal_id=1, card_id=1, is_online=True, is_fraud=False),
        Transaction(90, datetime(2023, 1, 15), terminal_id=0, card_id=0, is_online=False, is_fraud=False),
        Transaction(210, datetime(2023, 1, 20), terminal_id=1, card_id=1, is_online=True, is_fraud=False),
        Transaction(130, datetime(2023, 1, 25), terminal_id=0, card_id=0, is_online=False, is_fraud=False),
        # Training data
        Transaction(170, datetime(2023, 2, 1), terminal_id=1, card_id=1, is_online=True, is_fraud=False),
        Transaction(160, datetime(2023, 2, 5), terminal_id=0, card_id=0, is_online=False, is_fraud=False),
        # Test transaction
        Transaction(140, datetime(2023, 3, 10), terminal_id=1, card_id=1, is_online=True, is_fraud=False),
    ]
)
system = Banksys(
    transactions,
    cards,
    terminals,
    aggregation_windows=(timedelta(days=1), timedelta(days=7), timedelta(days=30)),
    clf_params=ClassificationParameters(training_duration=timedelta(days=30)),
    attackable_terminal_factor=1.0,
    fp_rate=0,
    fn_rate=0,
)
features = system.make_features(transactions)

# Retrieve the next transaction to perform in the system
data = system._transactions_df[-1].to_dicts()[0]
trx = Transaction(**data)

system.make_features()
```
