### Create Matrices

```python
classifiers = ['DNN', 'RF']
datasets = ['Kaggle', 'Generator']
balance_levels = ['balance=0.1', 'balance=0.5']
```

```python
import pandas as pd
metrics_df = pd.DataFrame(columns=['accuracy', 'precision', 'recall', 'f1'])

for clasifier in classifiers:
    for dataset in datasets:
        for balance in balance_levels:
            file_to_open = os.path.join(folder, clasifier, dataset, balance, 'metrics.json')
            with open(file_to_open, 'r') as file:
                data = json.load(file)
                print(data)
            metrics_df.loc[f'{clasifier}_{dataset}_{balance}', :] = [data['accuracy'], data['precision'], data['recall'], data['f1']]
```
