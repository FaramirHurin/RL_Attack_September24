```python
import pandas as pd
import matplotlib.pyplot as plt

n_clusters = '1'
class_sep = '1'
# 
# Load the CSV file
file_path = 'averaged_results/n_features=32_n_clusters='+n_clusters+'_class_sep='+class_sep+'_balance=0.1_k-4_u-6_fraudulent_reward-probability.csv'
data = pd.read_csv(file_path)

# Set the moving average window size (adjust as needed)
window_size = 100
moving_average = data.rolling(window=window_size).mean()

# Plot each column's moving average with its name as the label
plt.figure(figsize=(10, 6))

# Loop through each column and plot its moving average
for column in moving_average.columns:
    plt.plot(moving_average[column], label=column)

# Add labels and legend
plt.xlabel('Reward')
plt.ylabel('Frauds')
plt.title(f'Reward over time. N_clusters = ' + str(n_clusters) + ' class sep = ' + str(class_sep))
plt.legend()
plt.show()
```

```python
import os
import pandas as pd

# Set the directory path
folder_path = 'averaged_results'

# Get a list of all files in the folder (assuming they are CSV files)
file_list = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Initialize an empty list to store the dataframes
df_list = []

# Load each dataframe and add it to the list
for file in file_list:
    df = pd.read_csv(os.path.join(folder_path, file))
    df_list.append(df)

# Calculate the average of all dataframes by summing and then dividing by the number of dataframes
average_df = sum(df_list) / len(df_list)

# Optionally, save the averaged dataframe to a new file
average_df.to_csv(os.path.join(folder_path, 'average_results.csv'), index=False)


```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import re

# Set the directory path
folder_path = 'averaged_results'

# Initialize dictionaries to hold DataFrames by n_clusters
reward_probability_dfs = {}
reward_label_dfs = {}

# Get lists of files for each type and group them by n_clusters
for file in os.listdir(folder_path):
    if file.endswith('reward-probability.csv'):
        n_clusters = re.search(r'n_clusters=(\d+)', file).group(1)
        df = pd.read_csv(os.path.join(folder_path, file))
        if n_clusters not in reward_probability_dfs:
            reward_probability_dfs[n_clusters] = []
        reward_probability_dfs[n_clusters].append(df)

    elif file.endswith('reward-label.csv'):
        n_clusters = re.search(r'n_clusters=(\d+)', file).group(1)
        df = pd.read_csv(os.path.join(folder_path, file))
        if n_clusters not in reward_label_dfs:
            reward_label_dfs[n_clusters] = []
        reward_label_dfs[n_clusters].append(df)

# Function to average and plot moving averages for each n_clusters
def average_and_plot_for_n_clusters(n_clusters, reward_probability_dfs, reward_label_dfs):
    # Average DataFrames for the current n_clusters
    if n_clusters in reward_probability_dfs:
        average_prob_df = sum(reward_probability_dfs[n_clusters]) / len(reward_probability_dfs[n_clusters])
        moving_average_prob = average_prob_df.rolling(window=5).mean()  # Adjust window size as needed
        
        plt.figure(figsize=(10, 5))
        for column in moving_average_prob.columns:
            plt.plot(moving_average_prob[column], label=f'{column}')
        
        plt.title(f'Moving Average of Reward Probability (n_clusters={n_clusters})')
        plt.xlabel('Index')  # Change this to your actual x-axis label if applicable
        plt.ylabel('Value')
        plt.legend()
        plt.grid()
        plt.show()

    # Average DataFrames for the current n_clusters
    if n_clusters in reward_label_dfs:
        average_label_df = sum(reward_label_dfs[n_clusters]) / len(reward_label_dfs[n_clusters])
        moving_average_label = average_label_df.rolling(window=5).mean()  # Adjust window size as needed
        
        plt.figure(figsize=(10, 5))
        for column in moving_average_label.columns:
            plt.plot(moving_average_label[column], label=f'{column}')
        
        plt.title(f'Moving Average of Reward Label (n_clusters={n_clusters})')
        plt.xlabel('Index')  # Change this to your actual x-axis label if applicable
        plt.ylabel('Value')
        plt.legend()
        plt.grid()
        plt.show()

# Generate plots for each n_clusters group
for n_clusters in set(reward_probability_dfs.keys()).union(set(reward_label_dfs.keys())):
    average_and_plot_for_n_clusters(n_clusters, reward_probability_dfs, reward_label_dfs)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import re

# Set the directory path
folder_path = 'averaged_results'

# Initialize dictionaries to hold DataFrames by n_clusters
reward_probability_dfs = {}
reward_label_dfs = {}

# Get lists of files for each type and group them by n_clusters
for file in os.listdir(folder_path):
    if file.endswith('reward-probability.csv'):
        n_clusters = re.search(r'n_clusters=(\d+)', file).group(1)
        df = pd.read_csv(os.path.join(folder_path, file))
        if n_clusters not in reward_probability_dfs:
            reward_probability_dfs[n_clusters] = []
        reward_probability_dfs[n_clusters].append(df)

    elif file.endswith('reward-label.csv'):
        n_clusters = re.search(r'n_clusters=(\d+)', file).group(1)
        df = pd.read_csv(os.path.join(folder_path, file))
        if n_clusters not in reward_label_dfs:
            reward_label_dfs[n_clusters] = []
        reward_label_dfs[n_clusters].append(df)

# Function to plot moving averages for all n_clusters in a composed figure
def plot_composed_figure(reward_probability_dfs, reward_label_dfs):
    n_clusters_list = sorted(set(reward_probability_dfs.keys()).union(set(reward_label_dfs.keys())))
    
    # Determine the layout of subplots
    num_clusters = len(n_clusters_list)
    fig, axs = plt.subplots(1, num_clusters, figsize=(14, 5 * num_clusters))
    
    for idx, n_clusters in enumerate(n_clusters_list):
        # Plot moving averages for reward probabilities
        if n_clusters in reward_probability_dfs:
            average_prob_df = sum(reward_probability_dfs[n_clusters]) / len(reward_probability_dfs[n_clusters])
            moving_average_prob = average_prob_df.rolling(window=5).mean()  # Adjust window size as needed
            
            for column in moving_average_prob.columns:
                axs[idx, 0].plot(moving_average_prob[column], label=f'{column}')
            
            axs[idx, 0].set_title(f'Moving Average of Reward Probability (n_clusters={n_clusters})')
            axs[idx, 0].set_xlabel('Index')  # Change this to your actual x-axis label if applicable
            axs[idx, 0].set_ylabel('Value')
            axs[idx, 0].legend()
            axs[idx, 0].grid()



    plt.tight_layout()
    plt.show()

# Generate the composed figure for all n_clusters
plot_composed_figure(reward_probability_dfs, reward_label_dfs)
```
