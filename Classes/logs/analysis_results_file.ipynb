```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Base directory where the CSV files are stored
base_dir = '2024-10-14-17-17-17/Kaggle/'

# Parameters for rolling mean
window_size = 100

# Directories to average over
dirs_to_average = ['_0', '_1']

# Initialize dictionaries to store dataframes for each combination of k and u, separated by reward type
df_dict_label = {}
df_dict_prob = {}

# Iterate over all possible directories and files for different values of k and u
for k in [4]:  # Replace with actual values of k if different
    for u in [6]:  # Replace with actual values of u if different
        # Initialize lists to store dataframes for both reward types (label and probability)
        dfs_label = []
        dfs_prob = []
        
        for dir_suffix in dirs_to_average:
            # Path for reward-label
            path_label = f'{base_dir}/{dir_suffix}/balance=0.5/0/k-{k}/u-{u}/fraudulent/reward-label/'
            
            # Path for reward-probability
            path_prob = f'{base_dir}/{dir_suffix}/balance=0.5/0/k-{k}/u-{u}/fraudulent/reward-probability/'
            
            # List CSV files in reward-label directory
            csv_files_label = [f for f in os.listdir(path_label) if f.endswith('.csv')]
            for csv_file in csv_files_label:
                # Load the CSV file and append the dataframe to the list
                df_label = pd.read_csv(os.path.join(path_label, csv_file))
                dfs_label.append(df_label.iloc[:, 1:])  # Exclude first column if needed
            
            # List CSV files in reward-probability directory
            csv_files_prob = [f for f in os.listdir(path_prob) if f.endswith('.csv')]
            for csv_file in csv_files_prob:
                # Load the CSV file and append the dataframe to the list
                df_prob = pd.read_csv(os.path.join(path_prob, csv_file))
                dfs_prob.append(df_prob.iloc[:, 1:])  # Exclude first column if needed
        
        # Cell-by-cell average for reward-label
        if dfs_label:
            df_avg_label = sum(dfs_label) / len(dfs_label)  # Element-wise average
            df_dict_label[(k, u)] = df_avg_label
        
        # Cell-by-cell average for reward-probability
        if dfs_prob:
            df_avg_prob = sum(dfs_prob) / len(dfs_prob)  # Element-wise average
            df_dict_prob[(k, u)] = df_avg_prob

# Plot for reward-label
plt.figure(figsize=(10, 6))
for (k, u), df_avg_label in df_dict_label.items():
    # Calculate rolling mean for the averaged dataframe
    rolling_df_label = df_avg_label.rolling(window=window_size).mean()
    
    # Plot the rolling mean
    plt.plot(rolling_df_label)
plt.title(f'Rolling Mean for Reward-Label (Window Size = {window_size})')
plt.xlabel('Index')
plt.ylabel('Rolling Mean Value')
plt.legend(df_avg_label.columns, loc='upper left')
plt.show()

# Plot for reward-probability
plt.figure(figsize=(10, 6))
for (k, u), df_avg_prob in df_dict_prob.items():
    # Calculate rolling mean for the averaged dataframe
    rolling_df_prob = df_avg_prob.rolling(window=window_size).mean()
    
    # Plot the rolling mean
    plt.plot(rolling_df_prob)
plt.title(f'Rolling Mean for Reward-Probability (Window Size = {window_size})')
plt.xlabel('Index')
plt.ylabel('Rolling Mean Value')
plt.legend(df_avg_prob.columns, loc='upper left')
plt.show()
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Define root directory where the datasets are located
root_dir = '2024-10-18-11-21-49'
output_dir = 'Graphs'  # Subfolder to store the graphs

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Parameters for grouping
k_values = ['k-0', 'k-2', 'k-6']  # Example of k values, adjust accordingly
u_values = ['u-0', 'u-2', 'u-6']  # Example of u values, adjust accordingly
types = ['reward-label', 'reward-probability']
prefixes = ['_0', '_1', '_2', '_3']  # Prefixes to merge over

# Function to read and merge files for each combination of k, u, and type
def read_and_merge_files(k, u, data_type):
    sum_df = None
    count = 0

    # Dynamically find all timestamped folders
    timestamp_folders = [f for f in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, f))]

    for prefix in prefixes:
        for timestamp in timestamp_folders:
            folder_path = f'{root_dir}/Kaggle/{prefix}/balance=0.1/3/{k}/{u}/fraudulent/{data_type}/'
            file_path = os.path.join(folder_path, 'file.csv')  # Define the file path

            # Check if the folder exists
            if os.path.exists(file_path):
                df = pd.read_csv(file_path)
                if sum_df is None:
                    sum_df = df.copy()  # Initialize sum_df with the first dataframe
                else:
                    sum_df += df  # Element-wise addition
                count += 1

    if count > 0:
        average_df = sum_df / count  # Element-wise average
        return average_df
    else:
        return pd.DataFrame()  # Return empty dataframe if no files are found

# Apply rolling mean and plot
def plot_rolling_mean(merged_df, window_size=100, title="Rolling Mean Plot", save_path=None):
    rolling_mean_df = merged_df.rolling(window=window_size).mean()
    
    plt.figure(figsize=(10, 6))
    rolling_mean_df.plot()
    plt.title(f'{title} (Window={window_size})')
    plt.xlabel('Index')
    plt.ylabel('Values')
    
    # Save the plot if a save path is provided
    if save_path:
        plt.savefig(save_path)
    
    plt.show()

# Iterate over combinations of k, u, and type, merge the datasets, and plot
for k in k_values:
    for u in u_values:
        for data_type in types:
            merged_df = read_and_merge_files(k, u, data_type)
            if not merged_df.empty:
                plot_title = f'k={k}, u={u}, type={data_type}'
                # Construct the save path for the graph
                save_filename = f'{output_dir}/k_{k}_u_{u}_{data_type}.png'
                plot_rolling_mean(merged_df, window_size=100, title=plot_title, save_path=save_filename)
```
